```python
## WEB BASED LOADER WITH BEAUTIFULSOUP PARSING

from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_path="https://en.wikipedia.org/wiki/List_of_highest-grossing_Indian_films",
                       bs_kwargs=dict(parse_only = bs4.SoupStrainer(
                           class_ =("mw-heading mw-heading2")
                       ))
                        )

data = loader.load()
data # Display the first 500 characters of the content
```
